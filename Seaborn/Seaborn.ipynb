## Visualizing Statistical Relationship

```python
# relplot() makes use of two other axes functions for visualizing stastical relationships which are scatterplot() and lineplot()
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='line')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='scatter',hue='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='scatter',hue='smoker',style='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='scatter',hue='smoker',col='time')
plt.show()
```

### More on Line Plots

# Plotting with Categorical Data

```python
# catplot()  for different types of categorical values in seaborn

# Categorical scatterplots 
  # stripplot() Default  one of the variables is categorical
  # swarmplot() It does n't allow overlapping of markers

# Categorical distribution plots
    #boxplot() We can find min max median values
    #violinplot() 
    #boxenplot() plot different quarter values

# Categorical Estimate plots:
    #pointplot()
    #barplot()
    #countplot()

    
# Default categorical data is represented in the form 
```

```python
df=sns.load_dataset('tips')
df.head(3)
```

```python
sns.catplot(x='time',y='total_bill',data=df,jitter=0.3)
plt.show()
```

```python
sns.catplot(x='time',y='total_bill',data=df,kind='swarm')
plt.show()
```

```python
sns.catplot(x='time',y='total_bill',data=df,kind='swarm',hue='sex')
plt.show()
```

```python
sns.catplot(x='time',y='total_bill',data=df,kind='box')
plt.show()
```

```python
sns.catplot(x='time',y='total_bill',data=df,kind='boxen')
plt.show()
```
