```python
from bs4 import BeautifulSoup

data=BeautifulSoup(driver.page_source)
data
```

```python
ans=data.find(class_='q-box qu-borderAll qu-borderRadius--small qu-borderColor--raised qu-boxShadow--small qu-bg--raised')
len(ans)
ans
```

```python
ans1=ans.find_all(class_='q-box qu-borderBottom qu-p--medium')
len(ans1)
```

```python
for i in ans1:
    j=i.stripped_strings
    for k in j:
        if(k=='Follow'):
            break
        print(k,end='')
    print()
```
