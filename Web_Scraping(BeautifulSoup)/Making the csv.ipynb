```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get("http://books.toscrape.com/")
html=r.text
data=BeautifulSoup(html,'html.parser')
data
```

```python
# making links for all pages
base_url='http://books.toscrape.com/'
i=1
links=[]
while i != 50:
    links.append(base_url+'catalogue/page-'+str(i)+'.html')
    i+=1
```

```python
temp_data=data.find_all('article',class_='product_pod')
temp1=[]
for i in temp_data:
    temp2=[]
    title_name=i.h3.a.get("title")
    price=i.find(class_="price_color").string
    price=price[1:]
#     price1=price.p.name
    link=i.h3.a.get('href')
    link="http://books.toscrape.com/"+link
    temp2.append(title_name)
    temp2.append(price)
    temp2.append(link)
    temp1.append(temp2)
temp1
```

```python
Data=[]
for i in links:
    r=requests.get(i)
    html=r.text
    data=BeautifulSoup(html,'html.parser')
    temp_data=data.find_all('article',class_='product_pod')
    temp1=[]
    for i in temp_data:
        temp2=[]
        title_name=i.h3.a.get("title")
        price=i.find(class_="price_color").string
        price=price[1:]
        link=i.h3.a.get('href')
        link="http://books.toscrape.com/"+link
        temp2.append(title_name)
        temp2.append(price)
        temp2.append(link)
        Data.append(temp2)
Data
```

```python
import pandas as pd
df = pd.DataFrame(Data, columns = ['Title', 'Price', 'Link'])
df
```

```python
df.to_csv('books.csv', index = False)
```
