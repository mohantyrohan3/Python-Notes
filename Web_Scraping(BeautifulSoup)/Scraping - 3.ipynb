```python
from bs4 import BeautifulSoup
import requests
```

```python
r=requests.get("http://books.toscrape.com/")
r
```

```python
html=r.text
data=BeautifulSoup(html,'html.parser')
```

```python
data.head
```

```python
data.title
```

```python
# Details of First book

ans=data.find_all('article',class_='product_pod')
ans
```

```python
# First Book
Title1=ans[0].h3.a.get('title')
Title1
```

```python
# Finding Href of each and every book present in first page

for i in ans:
    print("http://books.toscrape.com/"+i.h3.a.get("href"))
```

```python
# Finding all the categories

cat=data.find('ul',class_='nav nav-list')
ans1=cat.ul.find_all('li')
for i in ans1:
    temp=list(i.a.stripped_strings)
    print(temp[0])
```
