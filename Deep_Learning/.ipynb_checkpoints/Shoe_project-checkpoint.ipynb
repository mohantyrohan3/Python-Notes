```python
import matplotlib.pyplot as plt
import cv2 as cv
import tensorflow as tf
from tensorflow.keras import layers,models
import numpy as np
```

```python
img=plt.imread('Shoe vs Sandal vs Boot Dataset\\Boot\\boot (2).jpg')
img.shape
```

```python
plt.imshow(img)
```

```python
dataset=tf.keras.utils.image_dataset_from_directory(
    "Shoe vs Sandal vs Boot Dataset",
    labels="inferred",
    label_mode="int",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64,64),
)
```

```python
classes=dataset.class_names
classes
```

```python
for image,label in dataset.take(1):
    plt.imshow(image[0].numpy().astype('uint8'))
    print(classes[label[0].numpy()])
```

```python
# Train test split
len(dataset)
train_size=0.8
train_s=len(dataset)*train_size
train_s
```

```python
train=dataset.take(375)
test=dataset.skip(375)
```

```python
val_size=0.1
len(dataset)*val_size
```

```python
validation=test.take(47)
len(validation)
```

```python
test=test.skip(47)
len(test)
```

```python
len(test)
```

```python
resize_and_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(64,64),
    layers.experimental.preprocessing.Rescaling(1.0/255)
])
```

```python
model = models.Sequential([resize_and_rescale])
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(3))
model.build(input_shape=(32,64,64,3))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train,epochs=10,
                    batch_size=32,
                    validation_data=validation)
```

# Prediction

```python
for image_batch,label_batch in test.take(1):
    plt.imshow(image_batch[0].numpy().astype('uint8'))
    #print("Actual Image :  "," ",classes[label[0].numpy()])
    
    
    batch_prediction = model.predict(image_batch)
    print("predicted label:",classes[np.argmax(batch_prediction[0])])
```
