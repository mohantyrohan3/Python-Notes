```python
Sample='Does This thing really work? Lets see'
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
sent_tokenize(Sample)
```

```python
words=word_tokenize(Sample.lower())
```

```python
from nltk.corpus import stopwords
```

```python
stop=stopwords.words('english')
stop
```

```python
word=[]
for i in words:
    if i not in stop:
        word.append(i)
word
```

```python
words
```

```python
import string
punctuations=list(string.punctuation)
punctuations
```

```python
word=[]
for i in words:
    if (i not in stop) & (i not in punctuations):
        word.append(i)
word
```

## Stemming

```python
stem_words=['play','played','playing','player','history','briefly']
ps=PorterStemmer()
```
