## Exploring json data

```python
response=requests.get('https://codingninjas.in/api/v3/courses')
print(response.status_code)
```

```python
import json
python_data=json.loads(response.text)
print(python_data)
```

```python
allcourses=python_data["data"]["courses"]
print(type(allcourses))
```

```python
# Name of all courses
for i in allcourses:
    print(i["title"])
```

## Passing Parameters


```python
a=requests.get('https://api.openaq.org/v2/cities?limit=100&page=1&offset=0&sort=asc&order_by=city',params={"country":"AU"})
a.status_code
p=json.loads(a.text)
p
# We can pass a number of parameters along with get request
# requests.get(url,params={ Key value pairs  })
```

```python
ans=p["results"]
len(ans)
```
