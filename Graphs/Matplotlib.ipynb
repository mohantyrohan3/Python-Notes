## Customizing Graphs

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.linspace(-10,10,1000)
y=x**3
y2=x**2
#plt.plot(x,y,color="black", marker="o" )
plt.plot(x,y,color="black",linewidth=0.5,label="X^3")
plt.plot(x,y2,color="red",linewidth=0.5,label="X^2")

#plt.axis([0,10,0,100]) # First 2 are for x axis and second two are for Y axis


plt.ylabel("Y")
plt.xlabel("X")

plt.grid()

plt.legend()
# plt.text(-2.5,0,"Intersection Point",#fontsize)

#plt.title("MAtplotlib Demo")
plt.show()
```

## Bubble Chart

```python
## Bubble Chart is used for representing 2 charts at a time in single chart
import matplotlib.pyplot as plt
year=[2014,2015,2016,2018,2019]
salary=[12,13,14,15,16]
population=[100,120,160,200,240]
```

```python
plt.scatter(year,salary,s=population,c="r") # s states for size
plt.title("Bubble Chart")
plt.show()
```

```python
import numpy as np
co=np.arange(len(year))
plt.scatter(year,salary,s=population,c=co,alpha=0.5,edgecolor="black")
plt.title("Bubble Chart")
plt.show()
```

```python
import numpy as np
co=np.arange(len(year))
plt.scatter(year,salary,s=population,c=co,alpha=0.5,edgecolor="black",marker="^")
plt.title("Bubble Chart")
plt.text(year[0]+0.05,salary[0]+0.05,population[0])
plt.show()
```

## Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
labels=["A","B","C","D"]
sizes=[4,3,6,9]
colors=["blue","purple","red","pink"]
explode=[0,0.1,0,0]
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="%.1f%%")  # counterclock=False , startangle=100
plt.title("Split Among Classes")
plt.axis("equal")
plt.show()
```
